```python
# Dependencies and Setup
import pandas as pd
import numpy as np

# File to Load
purchase_csv = "Resources/purchase_data.csv"

# Read Purchasing File and store into Pandas data frame
purchase_data = pd.read_csv(purchase_csv)
purchase_data.head()
```

## Player Count

* Display the total number of players


```python
# calculate total number of unique players
total_players = purchase_data["SN"].unique()
print(f"Total Number of Players = {len(total_players)}")
```

## Purchasing Analysis (Total)

* Run basic calculations to obtain number of unique items, average price, etc.


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame


```python
# calculate number of unique items purchased
unique_items = purchase_data["Item Name"].unique()
total_unique_items = len(unique_items)

# calculate average purchase price
avg_price = purchase_data["Price"].mean()

# calculate total number of purchases
total_purchase = purchase_data["Purchase ID"].count()

# calculate total revenue
total_revenue = purchase_data["Price"].sum()
```

```python
# create summary data frame
summary_df = pd.DataFrame({"Number of Unique Items":[total_unique_items],
                          "Average Price":[avg_price],
                          "Number of Purchases":total_purchase,
                          "Total Revenue":total_revenue
                          })

# format values in summary data frame
summary_df["Average Price"] = summary_df["Average Price"].map("${:.2f}".format)
summary_df["Total Revenue"] = summary_df["Total Revenue"].map("${:,.2f}".format)

# display formatted summary data frame
summary_df
```

## Gender Demographics

* Percentage and Count of Male Players


* Percentage and Count of Female Players


* Percentage and Count of Other / Non-Disclosed




```python
# drop duplicate player names to get a new data frame of unique players
cleaned_data = purchase_data.drop_duplicates(subset=["SN"])
cleaned_data.head()
```

```python
# get unique value counts for gender column and storing it in a new series
gender_count = cleaned_data["Gender"].value_counts()
gender_count
```

```python
# calculate count and percentage of male players
male_players = gender_count.loc["Male"]
percent_male_players = male_players/len(total_players)*100

# calculate count and percentage of female players
female_players = gender_count.loc["Female"]
percent_female_players = female_players/len(total_players)*100

# calculate count and percentage of other or non-disclosed players
other_players = gender_count.loc["Other / Non-Disclosed"]
percent_other_players = other_players/len(total_players)*100
```

```python
# create and format gender summary table based on the calculations and set index
gender_summary = pd.DataFrame({"Total Count": [male_players, female_players, other_players], 
                               "Percentage of Players": [percent_male_players, percent_female_players, percent_other_players]}, 
                                index = ["Male", "Female", "Other / Non-Disclosed"])

# format values in summary data frame
gender_summary["Percentage of Players"] = gender_summary["Percentage of Players"].map("{:.2f}%".format)

# display formatted summary data frame
gender_summary
```


## Purchasing Analysis (Gender)

* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. by gender




* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
# grouping purchase_data dataframe by Gender
purchase_by_gender = purchase_data.groupby("Gender")

# calculate purchase count
purchase_count = purchase_by_gender["Purchase ID"].count()

# calculate average purchase price
avg_purchase_price = purchase_by_gender["Price"].mean()

# calculate total purchase
total_purchase = purchase_by_gender["Price"].sum()

# calculate average purchase per person
avg_total_purchase_per_person = total_purchase/gender_count

print(purchase_count)
print(avg_purchase_price)
print(total_purchase)
print(avg_total_purchase_per_person)
```

```python
# create purchase analysis summary table by gender
purchase_analysis_gender = pd.DataFrame({"Purchase Count": purchase_count,
                                         "Average Purchase Price":avg_purchase_price,
                                         "Total Purchase Value": total_purchase,
                                         "Avg Total Purchase per Person":avg_total_purchase_per_person})

# rename index as Gender
purchase_analysis_gender = purchase_analysis_gender.rename_axis('Gender')

# format values in purchase analysis summary table
purchase_analysis_gender["Average Purchase Price"] = purchase_analysis_gender["Average Purchase Price"].map("${:.2f}".format)
purchase_analysis_gender["Total Purchase Value"] = purchase_analysis_gender["Total Purchase Value"].map("${:,.2f}".format)
purchase_analysis_gender["Avg Total Purchase per Person"] = purchase_analysis_gender["Avg Total Purchase per Person"].map("${:.2f}".format)

#display purchase analysis summary table
purchase_analysis_gender
```

## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python
# define bins and labels
age_bins = [0,9,14,19,24,29,34,39,50]
age_labels = ["<10","10-14","15-19","20-24","25-29","30-34","35-39","40+"]
```

```python
# perform binning on age column in cleaned_data dataframe and add labels to existing data frame as a new column
cleaned_data["Labels"] = pd.cut(cleaned_data["Age"], age_bins, labels = age_labels)
cleaned_data.head()
```

```python
# groupby labels and set them as index
age_index = cleaned_data.groupby("Labels")

# calculate total count age in each bin
total_count_age = age_index["Purchase ID"].count()

# calculate percentage of players in each bin
percent_players = total_count_age/len(total_players)*100

print(total_count_age)
print(percent_players)
```

```python
# create age demographics table
age_demographics = pd.merge(total_count_age, percent_players, on="Labels")

# rename columns in the dataframe
age_demographics = age_demographics.rename(columns={
    "Purchase ID_x":"Total Count", 
    "Purchase ID_y":"Percentage of Players"})


# format value in age demographics table
age_demographics["Percentage of Players"] = age_demographics["Percentage of Players"].map("{:.2f}%".format)

# remove index name
age_demographics.index.name = None

#display age demographics table
age_demographics
```

## Purchasing Analysis (Age)

* Bin the purchase_data data frame by age


* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. in the table below


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
# perform binning on age column in purchase_data dataframe and add labels to existing data frame as a new column
purchase_data["Labels"] = pd.cut(purchase_data["Age"], age_bins, labels = age_labels)
purchase_data.head()
```

```python
# groupby labels and set them as index
purchase_data_by_age = purchase_data.groupby("Labels")
purchase_data_by_age
```

```python
# calculate purchase count
purchase_count_age = purchase_data_by_age["Purchase ID"].count()

# calculate average purchase price
avg_purchase_price_age = purchase_data_by_age["Price"].mean()

# calculate total purchase
total_purchase_age = purchase_data_by_age["Price"].sum()

# calculate average purchase per person by dividing total purchase in each bin by unique total count in each bin
avg_total_purchase_per_person_age = total_purchase_age/total_count_age

print(purchase_count_age)
print(avg_purchase_price_age)
print(total_purchase_age)
print(avg_total_purchase_per_person_age)
```

```python
# merge purchase count age and average purchase price dataframe into a new dataframe
purchase_analysis_by_age = pd.merge(purchase_count_age, avg_purchase_price_age, on="Labels")

# merge total purchase dataframe into purchase analysis by age dataframe
purchase_analysis_by_age = pd.merge(purchase_analysis_by_age, total_purchase_age, on="Labels")

# add average total purchase per person age series as the column in dataframe
purchase_analysis_by_age ["Avg Total Purchase per Person"] = avg_total_purchase_per_person_age

# rename columns
purchase_analysis_by_age = purchase_analysis_by_age.rename(columns={
    "Purchase ID":"Purchase Count", 
    "Price_x":"Average Purchase Price", 
    "Price_y":"Total Purchase Value"})

# rename axis
purchase_analysis_by_age = purchase_analysis_by_age.rename_axis('Age Ranges')

# format values in purchase analysis summary table
purchase_analysis_by_age["Average Purchase Price"] = purchase_analysis_by_age["Average Purchase Price"].map("${:.2f}".format)
purchase_analysis_by_age["Total Purchase Value"] = purchase_analysis_by_age["Total Purchase Value"].map("${:,.2f}".format)
purchase_analysis_by_age["Avg Total Purchase per Person"] = purchase_analysis_by_age["Avg Total Purchase per Person"].map("${:.2f}".format)

# display purchase analysis by age summary table
purchase_analysis_by_age
```
